```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = cv.imread('images/emma.jpg',cv.IMREAD_GRAYSCALE)
fig,ax = plt.subplots(1,2,figsize=(5,5))
ax[0].imshow(f,cmap='gray')
t = np.arange(256,dtype=np.uint8)
#g = t[f]
g = cv.LUT(f,t)
ax[1].imshow(g,cmap='gray')
for a in ax:
    a.axis('off')
plt.show()
```

```python
# Negative Transformation
f = cv.imread('images/emma.jpg',cv.IMREAD_GRAYSCALE)
fig,ax = plt.subplots(1,2,figsize=(5,5))
ax[0].imshow(f,cmap='gray')
t = np.arange(255,-1,-1,dtype=np.uint8)
#print(t)
#g = t[f]
g = cv.LUT(f,t)
ax[1].imshow(g,cmap='gray')
for a in ax:
    a.axis('off')
plt.show()
```

```python
# Negative Transformation
f = cv.imread('images/Fig0304(a)(breast_digital_Xray).tif',cv.IMREAD_GRAYSCALE)
fig,ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(f,cmap='gray')
t = np.arange(255,-1,-1,dtype=np.uint8)
g = cv.LUT(f,t)
ax[1].imshow(g,cmap='gray')
for a in ax:
    a.axis('off')
plt.show()
```

```python
#Intensity Windowing

c = np.array([(100,50),(150,200)])

t1 = np.linspace(0,c[0,1],c[0,0]+1-0).astype('uint8')
#print(t1)

t2 = np.linspace(c[0,1]+1,c[1,1],c[1,0]-c[0,0]).astype('uint8')
#print(t2)

t3 = np.linspace(c[1,1]+1,255,255-c[1,0]).astype('uint8')
#print(t3)

transform = np.concatenate((t1,t2),axis=0).astype('uint8')
transform = np.concatenate((transform,t3),axis=0).astype('uint8')
#print(len(transform))

f = cv.imread('images/emma.jpg',cv.IMREAD_GRAYSCALE)
fig,ax = plt.subplots(1,2,figsize=(5,5))
ax[0].imshow(f,cmap='gray')

g = cv.LUT(f,transform)

ax[1].imshow(g,cmap='gray')
for a in ax:
    a.axis('off')
plt.show()

```

```python
f = cv.imread('Images/Fig0309(a)(washed_out_aerial_image).tif',cv.IMREAD_GRAYSCALE)
assert f is not None

gamma1 = 0.5
t1 = np.array([(i/255)**gamma1*255 for i in range(256)],dtype=np.uint8)
g1 = cv.LUT(f,t1)

gamma2 = 3
t2 = np.array([(i/255)**gamma2*255 for i in range(256)],dtype=np.uint8)
g2 = cv.LUT(f,t2)

fig,ax = plt.subplots(1,3,figsize=(10,10))
ax[0].imshow(f,cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(g1,cmap='gray')
ax[1].set_title(f'Transformed with $\gamma$ = {gamma1}')
ax[2].imshow(g2,cmap='gray')
ax[2].set_title(f'Transformed with $\gamma$ = {gamma2}')
for a in ax:
    a.axis('off')
plt.show()
```

```python
# Histrogram Equalization

f = cv.imread("Images/shells.tif",cv.IMREAD_GRAYSCALE)
g = cv.equalizeHist(f)

fig,ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(f,cmap='gray', vmin=0,vmax=255)
ax[0].set_title('Original')
ax[1].imshow(g,cmap='gray', vmin=0,vmax=255)
ax[1].set_title('Equalized')

for a in ax:
    a.axis('off')
plt.show()
```
