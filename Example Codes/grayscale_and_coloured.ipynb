```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
im = np.zeros((6,8),dtype=np.uint8)
im[2,3] = 255
fig,ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(im,cmap='gray',vmin=0,vmax=255)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
im2 = np.zeros((6,8,3),dtype=np.uint8)
im2[2,3] = (255,150,190)
fig,ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(im2)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
im3 = cv.imread('Images/emma.jpg')
fig,ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(cv.cvtColor(im3,cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
im3 = cv.imread('Images/emma.jpg')
fig,ax = plt.subplots(1,1,figsize=(6,8))
ax.imshow(cv.cvtColor(im3,cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
print(im3.size)
print(im3.shape)
print(im3.dtype)
```

```python
im4 = cv.imread('images/emma.jpg',cv.IMREAD_GRAYSCALE)
im5 = cv.add(im4,100)
fig,ax = plt.subplots(1,2,figsize=(6,8))
ax[0].imshow(im4,cmap='gray',vmin=0,vmax=255)
ax[0].set_title('Original Image')
ax[1].imshow(im5,cmap='gray',vmin=0,vmax=255)
ax[1].set_title('Brightness Increased Image')
for a in ax:
    a.axis('off')
plt.show()
```

```python
im6 = cv.imread('images/tiger.jpg')
im_blue = im6.copy()
im_blue[:,:,1] = 0
im_blue[:,:,2] = 0
fix,ax = plt.subplots(1,2,figsize=(12,8))
ax[0].imshow(cv.cvtColor(im6,cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[1].imshow(cv.cvtColor(im_blue,cv.COLOR_BGR2RGB))
ax[1].set_title('Blue Channel')
for a in ax:
    a.axis('off')
plt.show()
```
